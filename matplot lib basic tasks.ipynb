Q1: Read Total profit of all months and show it using a line plot
Total profit data provided for each month. Generated line plot must include the following properties: –
X label name = Month Number
Y label name = Total profit
The line plot graph should look like this.



```python
total_profit=np.random.randint(100000,500000,size=(12,))
months=np.arange(1,12+1)
print(total_profit)
plt.plot(months,total_profit)
plt.title("Company profit per month")
plt.xlabel("Month Number")
plt.ylabel("Total profit")
plt.xticks(months)
plt.show()

```

Q8. Calculate total sale data for last year for each product and show it using a Pie chart
Note: In Pie chart display Number of units sold per year for each product in percentage.
The Pie chart should look like this.



```python
categories = ['ToothPaste', 'FaceWash', 'FaceCream', 'Bathing Soap', 'Shampoo', 'Moisturizer']
random_sales = np.random.rand(len(categories))
random_sales = random_sales / random_sales.sum() * 100  

plt.pie(random_sales, labels=categories, autopct='%1.1f%%')
plt.title('Sales Data')
plt.legend([i for i in categories],loc=4)
plt.show()
```
